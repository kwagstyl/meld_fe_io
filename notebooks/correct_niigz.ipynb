```python
## correct nii.gz files that are not really zipped

# correct zip in .nii
import gzip
import shutil
import glob
import sys
import subprocess as sub

def check_gzip(file):
    # ensure file are gzip, or gzip
    if '.nii.gz' in file:
        try:
            with gzip.open(file, 'rb') as f:
                data = f.read(10)  # Read the first 10 bytes (adjust as needed)
        except OSError as e:
            if "Not a gzipped file" in str(e):
                print(f"ERROR: The file '{file}' has the suffix '.nii.gz. but is not genuinely compressed using gzip. Check that you have the right format and rerun")
                print('Zipp the file ')
                unzip_file = file.split('.nii.gz')[0] + '.nii'
                shutil.move(file, unzip_file)
                command = format(f'gzip {unzip_file}')
                sub.check_call(command, shell=True)
            elif "No such file" in str(e):
                print(f"ERROR: The file '{file}' was not found.")
                sys.exit()
            else:
                print(f"ERROR: An error occurred: {e}")
                sys.exit()
            
    else:
        try:
            with gzip.open(file, 'rb') as f:
                data = f.read(10)  # Read the first 10 bytes (adjust as needed)
            print(f"ERROR: The file '{file}' is compressed using gzip but does not have the right suffix '.nii.gz' ")
            sys.exit()
        except OSError as e:
            if "Not a gzipped file" in str(e):
                command = format(f'gzip {file}')
                sub.check_call(command, shell=True)
                file = file+'.gz'
    return file

# files = glob.glob('/home/mathilde/Documents/RDS/MELD_FE/DATA/MELD_H/MELD_BIDS/*/*/*.nii.gz')
files = glob.glob('/home/mathilde/Documents/RDS/MELD_FE/DATA/MELD_H104/raw_data/data_240219/MELD_beforeBIDS/*/*/*/*.nii.gz')

for file in files:
    print(file)
    check_gzip(file)
```
